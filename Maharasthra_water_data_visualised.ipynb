```python
Price_of_water = x.iloc[: , :1]
Price_of_water
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(8, 3))
plt.hist(
    Price_of_water['Price of water (Rs./Kl)'],
    bins=18,
    edgecolor='black',  # Set the border color to black
    histtype='step'     # No fill color inside the bins
)

plt.title('Price of water (Rs./Kl)')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

prices = Price_of_water.values.flatten()
consumption = total_consumption.values.flatten()

data_df = pd.DataFrame({
    'Price_of_Water': prices,
    'Total_Consumption': consumption
})

num_bins = 18
bins = np.linspace(data_df['Price_of_Water'].min(), data_df['Price_of_Water'].max(), num_bins + 1)

data_df['Price_Bin'] = pd.cut(data_df['Price_of_Water'], bins=bins)

bin_avg_consumption = data_df.groupby('Price_Bin')['Total_Consumption'].mean().reset_index()

bin_avg_consumption['Midpoint'] = bin_avg_consumption['Price_Bin'].apply(lambda x: x.mid)


plt.figure(figsize=(10, 5))
plt.plot(
    bin_avg_consumption['Midpoint'],
    bin_avg_consumption['Total_Consumption'],
    marker='x',
    linestyle='--',
    color='black' ,
    ms = 10
)
plt.title('Average Total Consumption vs. Price Intervals')
plt.xlabel('Price Interval Midpoint')
plt.ylabel('Average Total Consumption')
plt.grid(True)
plt.tight_layout()
plt.show()
```
